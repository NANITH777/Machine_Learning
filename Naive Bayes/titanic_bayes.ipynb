```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("D:\\Machine_Learning\\Machine_Learning\\Naive Bayes\\titanic.csv")
df
```

```python
df.columns[df.isna().any()]
```

```python
df.columns
```

```python
df.drop(['Cabin', 'Embarked', 'Name', 'SibSp','PassengerId','Parch','Ticket'], axis=1, inplace=True)
df.head()
```

```python
target=df.Survived
attributs=df.drop('Survived', axis=1)
attributs.head()
```

```python
dummy=pd.get_dummies(attributs.Sex).astype(int)
dummy.head()
```

```python
attributs=pd.concat([attributs, dummy], axis=1)
attributs.head()
```

```python
attributs=attributs.drop(['Sex'], axis=1)
attributs.head()
```

```python
attributs.columns[attributs.isna().any()]
```

```python
attributs.Age=attributs.Age.fillna(attributs.Age.mean())
attributs.head(10)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test =train_test_split(attributs, target, test_size=0.25)
```

```python
from sklearn.naive_bayes import GaussianNB
model=GaussianNB()
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```
