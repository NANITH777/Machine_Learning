```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
data=pd.read_csv('D:\Machine_Learning\Machine_Learning\multivariable regression\homeprices.csv')
data.head(5)
```

```python
import math
median_bedrooms=math.floor(data.bedrooms.median())
median_bedrooms
```

```python
data.bedrooms=data.bedrooms.fillna(median_bedrooms)
data.head()
```

```python
reg=linear_model.LinearRegression()
reg.fit(data[['area','bedrooms', 'age']], data.price)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.predict([[3000, 4, 40]])
```

```python
%matplotlib inline
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(data.area, data.price, color='blue', marker='+')
```

```python
%matplotlib inline
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(data.area, data.price, color='blue', marker='+')
plt.plot(data.area, reg.predict(data[['area','bedrooms','age']]), color='red')
```

```python
%matplotlib inline
plt.xlabel('bedrooms number')
plt.ylabel('price')
plt.scatter(data.bedrooms, data.price, color='blue', marker='+')
plt.plot(data.bedrooms, reg.predict(data[['area','bedrooms','age']]), color='red')
```

```python
%matplotlib inline
plt.xlabel('age')
plt.ylabel('price')
plt.scatter(data.age, data.price, color='blue', marker='+')
plt.plot(data.age, reg.predict(data[['area','bedrooms','age']]), color='red')
```
