```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
data=pd.read_csv('D:\Machine_Learning\Machine_Learning\linear regression\Canada Year capital\\canada_per_capita_income.csv')
data.head(5)
```

```python
plt.figure(figsize=(16, 10))
%matplotlib inline
plt.ylabel('year')
plt.xlabel('per capita income (US$)')
plt.scatter(data[['year']], data[['per capita income (US$)']], color='red', marker='*')
plt.title('year ve income')
```

```python
reg=linear_model.LinearRegression()
reg.fit(data[['year']], data[['per capita income (US$)']])
```

```python
reg.predict([[2020]])
```

```python
plt.figure(figsize=(16, 10))
%matplotlib inline
plt.ylabel('year')
plt.xlabel('per capita income (US$)')
plt.scatter(data[['year']], data[['per capita income (US$)']], color='red', marker='*')
plt.title('year ve income')
plt.plot(data.year, reg.predict(data[['year']]), color='blue')
```
